```python
import  numpy as np
import tensorflow as tf
from keras.models import load_model
from skimage.io import imread
from skimage.transform import resize
img_path = 'M6_P\AML_M6b_ID1523_PB.jpg'
# Cargar el modelo AlexNet desde un archivo h5
model = load_model('AlexNet_model.h5')
image = imread(img_path)
img = resize(image, (150, 250), anti_aliasing=True, preserve_range=True)
img_nor = img.astype('float32')
img_nor /= 255.0
img_array = tf.keras.preprocessing.image.img_to_array(img_nor)
img_array = tf.expand_dims(img_array, axis=0)
# Realizar la predicción
predictions1 = model.predict(img_array)
# Imprimir las predicciones
print(predictions1)
```

```python
indice_maximo = np.argmax(predictions1)
indice_maximo
```
